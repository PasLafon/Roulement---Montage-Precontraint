#### 2.1) Calcul des efforts sur les roulements :
Hypothèses classiques pour appliquer le PFD :
<ul>
    <li>Régime stationnaire (vitesse de rotation constante).</li>
    <li>Ensemble tournant (arbre+couronne dentées) parfaitement équilibré statiquement et dynamiquement.</li>
</ul>

<I>Ici si on modélise les roulements $R_1$ et $R_2$ comme des rotules (ou des demi-rotules) on supposera qu'il y a 3 inconnues d'efforts par roulement, soit 6 inconnues en tout. Le PFD permettra d'écrire 6 équations scalaires, mais il y a un degré de mobilité (rotation / axe $\vec{y}$), on a donc bien un système hyperstatique de degré 1.</I>

On isole  l'ensemble tournnant $S_2$, bilan des actions mécaniques extérieures :<br>

$$
\{\mathcal{T}_{S_1\to S_2}\}:
\left\lbrace
\begin{array}{c}
\vec{R}_I\\
\vec{0}
\end{array}
\right\rbrace_{I}
\quad
\{\mathcal{T}_{R_1\to S_2}\}:
\left\lbrace
\begin{array}{c}
\vec{R}_C\\
\vec{0}
\end{array}
\right\rbrace_{C}
\quad
\{\mathcal{T}_{R_2\to S_2}\}:
\left\lbrace
\begin{array}{c}
\vec{R}_D\\
\vec{0}
\end{array}
\right\rbrace_{D}
\quad
\{\mathcal{T}_{\text{Ext.}\to S_2}\}:
\left\lbrace
\begin{array}{c}
\vec{0}\\
\vec{\mathcal{C}}_\text{Mot}
\end{array}
\right\rbrace_{O}
$$

Le PFD en $D$ s'écrit :

$$
\left\lbrace
\begin{array}{c}
\vec{R}_I\\
\vec{DI}\wedge\vec{R}_I
\end{array}
\right\rbrace_{D}
+
\left\lbrace
\begin{array}{c}
\vec{R}_C\\
\vec{DC}\wedge\vec{R}_C
\end{array}
\right\rbrace_{D}
+
\left\lbrace
\begin{array}{c}
\vec{R}_D\\
\vec{0}
\end{array}
\right\rbrace_{D}
+
\left\lbrace
\begin{array}{c}
\vec{0}\\
\vec{\mathcal{C}}_\text{Mot}
\end{array}
\right\rbrace_{D}
=
\left\lbrace
\begin{array}{c}
\vec{0}\\
\vec{0}
\end{array}
\right\rbrace
$$

$$
\left\lbrace
\begin{array}{crcl}
0&-\frac{d_2}{2}&      &X_0\\
Y_0&L_2&\wedge&Y_0\\
Z_0&0     &      &Z_0
\end{array}
\right\rbrace_{D}
+
\left\lbrace
\begin{array}{crcl}
X_C&0&      &X_C\\
Y_C&L_1+L_2&\wedge&Y_C\\
Z_C&0     &      &Z_C
\end{array}
\right\rbrace_{D}
+
\left\lbrace
\begin{array}{rl}
X_D&0\\
Y_D&0\\
Z_D&0 
\end{array}
\right\rbrace_{D}
+
\left\lbrace
\begin{array}{rl}
0&0\\
0&\mathcal{C}_\text{Mot}\\
0&0
\end{array}
\right\rbrace_{D}
=
\left\lbrace
\begin{array}{rl}
0&0\\
0&0\\
0&0
\end{array}
\right\rbrace
$$

D'où le système de 6 équations :<br>

$$
\left\lbrace
\begin{array}{r}
X_C + X_D =0\\
Y_0 + Y_C + Y_D =0\\
Z_0 + Z_C + Z_D =0\\
L_2.Z_0+(L_1+L_2).Z_C =0\\
\frac{d_2}{2}.Z_0 + \mathcal{C}_\text{Mot}=0\\
-\frac{d_2}{2}.Y_0- (L_1+L_2).X_C =0
\end{array}
\right.
$$

Qui donne pour solutions :<br><br>
$$
X_C=-\frac{d_2}{2(L_1+L_2)}Y_0
\quad
Z_C=-\frac{L_2}{L_1+L_2}Z_0
\quad
X_D=\frac{d_2}{2(L_1+L_2)}Y_0
\quad
Z_D=-\frac{L_1}{L_1+L_2}Z_0
$$

<I>Conformément à l'analyse globale, les composantes $Y_C$ et $Y_D$ ne peuvent pas être déterminées car elles n'interviennent que dans la 2ième équations $Y_0+Y_C+Y_D=0$.</I>

Les efforts sur les roulements $R_1$ et $R_2$ sont :<br><br>
$$
\{\mathcal{T}_{S_2\to R_1}\}=-\{\mathcal{T}_{R_1\to S_2}\}:
\left\lbrace
\begin{array}{rl}
X_1=-X_C&0\\
Y_1=-Y_C&0\\
Z_1=-Z_C&0
\end{array}
\right\rbrace_{C}
\quad
\{\mathcal{T}_{S_2\to R_2}\}=-\{\mathcal{T}_{R_2\to S_2}\}:
\left\lbrace
\begin{array}{el}
X_2=-X_D&0\\
Y_2=-Y_D&0\\
Z_2=-Z_D&0
\end{array}
\right\rbrace_{D}
$$

Les efforts radiaux sont :<br><br>
$$
F_{r1}=\sqrt{X_1^2+Z_1^2}\quad\text{pour }R_1\qquad
F_{r2}=\sqrt{X_2^2+Z_2^2}\quad\text{pour }R_2
$$

```python
import numpy as np
Y0 = 2000
Z0 = -5700
d1 = 56
d2 = 160
L1 = 80
L2 = 40

X1 = 0.5*Y0*d2/(L1+L2)
Z1 = Z0*L2/(L1+L2)
Fr1 = np.sqrt(X1**2+Z1**2)

X2 = -X1
Z2 = Z0*L1/(L1+L2)
Fr2 = np.sqrt(X2**2+Z2**2)
print('Roult. R1 :\n\tX1  = %+.2f N\n\tZ1  = %+.2f N\n\tFr1 = %+.2f N'%(X1,Z1,Fr1))
print('Roult. R2 :\n\tX2  = %+.2f N\n\tZ2  = %+.2f N\n\tFr2 = %+.2f N'%(X2,Z2,Fr2))
```

#### 2.3) Durée de vie nécessaire par roulements :

<div class="alert alert-info">
Pour que les roulements puissent fonctionner dans des conditions acceptables, il faut qu'au minimum la charge axiale $F_a$ sur le roulement, soit telle qu'au moins <B>50% des éléments roulants soit chargés</B>.<br><br>
La charge axiale minimale dépend de :
<ul>
    <li>La charge radiale $F_r$ appliquée sur le roulement.</li>
    <li>La géométrie du roulement et le % d'éléments roulants chargés.</li>
</ul>
On doit vérifier la relation (propre aux roulements à rouleaux), avec $Y_2$ un coefficient dépendant du roulement :
$$\frac{F_a}{F_r}\geq\frac{0.5}{Y_2}$$
<br>
La durée de vie, en millions de tours, est calculée pour une fiabilité standard de 90% avec :
$$L_{10}=\left(\frac{C}{P}\right)^{\frac{10}{3}}$$
Avec :
<ul>
    <li>$P$ : La charge radiale équivalente appliquée sur le roulement en [N].</li>
    <li>$C$ : Capacité de charge dynamique du roulement en [N].</li>
</ul>
</div>

Pour éviter de tester tous les roulements on estime l'ordre de grandeur de la capacité dynamique nécessaire :
* On considère la plus forte charge radiale, ici $F_{r2}$ et on calcule : $C\simeq{L_{10}}^{3/10}F_{r2}$.

```python
print('Ordre de grandeur de la capacité dynamique %.f N:'%(L10**(3/10)*Fr2))
```

```python
## Fonction pour calculer la charge radiale équivalente :
def Peq(Fa,Fr,e,Y2):
    if (Fa/Fr)<=e:
        Peq = Fr
    else:
        Peq = 0.4*Fr + Y2*Fa
    return Peq
# Liste des roulements testés :
#          'ref'       :(d  D  Cr    e    Y_2  C)
CatRoul = {'4T-32007X' :(35,62,41500,0.45,1.32,14),
           '4T-33007'  :(35,62,50500,0.31,1.97,17),
           '4T-30207'  :(35,72,55500,0.37,1.60,15),
           '4T-32207'  :(35,72,72500,0.37,1.60,19),
           '4T-32207C' :(35,72,68000,0.58,1.03,19),
           '4T-32207CR':(35,72,62000,0.55,1.10,18),
           '4T-32008X' :(40,68,50000,0.38,1.58,14.5),
           '4T-33008'  :(40,68,59500,0.36,1.69,18)}

# Calcul des durées et des charges axiales nécessaires pour chaque
# roulement :
for ref in CatRoul:
    
    [d,D,Cr,e,Y2,c] = CatRoul[ref]
    
    # Roulement 1 -> le moins chargé axialement :
    Fa1 = 0.5*Fr1/Y2          # Charge axiale minimale pour 50% des éléments roulants chargés.
    Peq1 = Peq(Fa1,Fr1,e,Y2)  # Charge radiale équivalente correspondante.
    L101 = (Cr/Peq1)**(10/3)  # Durée de vie en millions de tours.
    
    # Roulement 2 -> le plus chargé axialement :
    Fa2min = 0.5*Fr2/Y2       # Charge axiale minimale pour 50% des éléments roulants chargés.
    Fa2 = Fa1 + np.abs(Y0)    # Charge axiale tirée  du PFD
    Peq2 = Peq(Fa2,Fr2,e,Y2)  # Charge radiale équivalente correspondante.
    L102 = (Cr/Peq2)**(10/3)  # Durée de vie en millions de tours.

    print(ref)
    print('R1 : Fa1 (50%%)        : %.2f N\t\tFr1 : %.2f N\tPeq1 : %.2f N   L10_1 : %.2f'%(Fa1,Fr1,Peq1,L101))
    print('R2 : Fa2 (PFD)/(50%%)  : %.2f/%.2f N\tFr2 : %.2f N\tPeq2 : %.2f N   L10_2 : %.2f'%(Fa2,Fa2min,Fr2,Peq2,L102))
    print(('-')*100)
 
```

### 3) Précontrainte minimale à imposer :

<div class="alert alert-info">
    <img src="ex-effort-montage-O.svg">
$$F_a=Z.\xi(a,t)\sin\alpha\left(\frac{\cos\alpha}{K_i+K_e}\right)^{\frac{1}{t}}\delta_r^{\frac{1}{t}}$$
$$F_r=Z.\psi(a,t)\cos\alpha\left(\frac{\cos\alpha}{K_i+K_e}\right)^{\frac{1}{t}}\delta_r^{\frac{1}{t}}$$
<ul>
    <li>$F_a$ : Effort axial sur le roulement.</li>
    <li>$F_r$ : Effort radial sur le roulement.</li>
    <li>$\delta_r$ : Déplacement radial de la bague intérieure/bague extérieure.</li>
    <li>$Z$ : nombre d'éléments roulants ($t=0.9$ pour les rouleaux, $t=2/3$ pour les billes).</li>
    <li>$K_i$,$K_e$ : Raideurs des éléments roulants, et pour des rouleaux :</li>
    <br>$$K_i+K_e=\frac{2\times 3.78\times 10^{-5}}{L^{0.8}}N^{-0.9}mm^{1.8}$$<br>
    <li>$a$ : Paramètre de charge, traduit le % d'éléments roulants chargés en fonctionnement. Pour $a=0$, 50% des éléments roulants chargés.</li>
    <li>$\alpha$ : Angle positionnant le centre de poussée $\to$ lié à la géométrie du roulement.</li>
    </ul>
Avec : $f(a,t)=\dfrac{\xi(a,t)}{\phi(a,t)}$ on a la relation :  
$$\frac{F_a}{F_r}=f(a,t).\tan\alpha$$
Les déplacements axial $\delta_a$ et radial $\delta_r$ sont tels que :
$$a=\frac{\delta_a}{\delta_r}\tan\alpha$$
Dans la plupart des catalogues, $\alpha$ n'est pas fournit, par contre les constructeurs donne la relation suivante pour $a=0$, donc 50% des éléments chargés :
$$\frac{F_a}{F_r}=f(0,t).\tan\alpha=\frac{0.5}{Y_2}$$    
</div>

Cas de l'exercice :<br>
On sait que :
<ol>
    <li>Les éléments roulants de $R_1$ sont chargés à 50%, donc le paramètre de charge de $R_1$, $a_1=0$, $\tan\alpha_1=\frac{0.5}{Y_2f(0,0.9)}$ et $\delta_{a1}=0$.</li>
    <li>Le roulement $R_2$ est identique à $R_1$ donc l'angle $\alpha_2=\alpha_1=\alpha$.</li>
    <li>Pour $R_2$ on a : $\frac{F_{a2}}{F_{r2}}=f(a_2,t)\tan\alpha$, donc connaissant $F_{a2}$, $F_{r2}$ et $\tan\alpha$ on pourra déterminer $f(a_2,t)$ et donc la valeur de $a_2$, puis $\delta_{r2}$ et $\delta_{a2}$.</li>
    </ol>

```python
## Calcul et Ttracé de la fonction f(a,t) :
import numpy as np
from scipy import integrate

import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
# Borne d'intégration :
def Borne(a):
    return  np.arccos(-min(a,1.0))

def Xi(a,t):
    dXi  = lambda x:(np.cos(x)+a)**(1/t)
    # Xi calculée par intégration numérique:
    Xi = 1/(2*np.pi)*integrate.quad(dXi,-Borne(a),Borne(a))[0]
    return Xi

def Psi(a,t):
    dPsi  =  lambda x:np.cos(x)*(np.cos(x)+a)**(1/t)
    # Phi calculée par intégration numérique:
    borne = Borne(a)
    Psi = 1/(2*np.pi)*integrate.quad(dPsi,-Borne(a),Borne(a))[0]
    return Psi


T=((2/3,'2/3'),(0.9,'0.9'))
# it = 0/1 -> billes/rouleaux
it=1
t= T[it][0]
tt = T[it][1]
ValXi = []
ValPsi = []
# Pour les valeurs de a entre -1 et 3
A = np.linspace(-1,2,200)
for a in A: 
    ValXi.append(Xi(a,t))
    ValPsi.append(Psi(a,t))          

# Tracé des courbes :
ValXi = np.array(ValXi)
ValPsi = np.array(ValPsi)
# Pour éviter le pb de la division par 0 pour a=-1.
f = np.full_like(ValPsi,1)
f[1:] = ValXi[1:]/ValPsi[1:]
plt.figure(figsize=(8,8))
plt.plot(A,ValXi,'r-',label=r'$\xi(a,$'+tt+'$)$')
plt.plot(A,ValPsi,'b-',label=r'$\psi(a,$'+tt+'$)$')
plt.plot(A,f,'g-',linewidth=3,label=r'$f(a,$'+tt+'$)$')
plt.xlim(-1,1)
plt.ylim(0,3)
plt.xlabel('Paramètre de charge $a$')
plt.xticks(np.arange(-1,2,0.2))
plt.yticks(np.arange(0,3,0.2))
plt.grid()
plt.legend(fontsize=12)
```

Calcul de l'angle $\alpha$ pour le roulement choisi `4T-32207` :

$$\frac{F_{a1}}{F_{r1}}=\frac{0.5}{Y_2}=f(0,0.9)\tan\alpha=\frac{\xi(0,0.9)}{\phi(0,0.9)}\tan\alpha\Rightarrow \alpha=\arctan\left(\frac{0.5}{Y_2f(0,0.9)}\right)$$

```python
# Recupère les caractéristiques du roulement :
[d,D,Cr,e,Y2,C] = CatRoul['4T-32207']
# Recalcule les valeurs des efforts axiaux pour ce roulement
Fa1 = 0.5*Fr1/Y2  
Fa2 = Fa1 + np.abs(Y0)  
f_a1 = Xi(0,0.9)/Psi(0,0.9)
alpha =np.arctan(0.5/(Y2*f_a1))
print('alpha : %.2f deg'%(alpha*180/np.pi))
```

Calcul de $f(a_2,t)$ pour $R_2$ :
$$\frac{F_{a2}}{F_{r2}}=f(a_2,t)\tan\alpha\Rightarrow f(a_2,t)=\frac{F_{a2}}{F_{r2}\tan\alpha }$$

```python
f_a2 = Fa2/(np.tan(alpha)*Fr2)
print('f_a2 : %.4f'%f_a2)
```

Maintenant il faut retrouver la valeur de $a_2$ donnant $f(a_2,t)=2.7244$.<br>
Deux solutions :
<ol>
    <li>Avec la courbe ci-dessus, on trouve $a_2\simeq1.5$.</li>

    </ol>

```python
# Résolution numérique avec une méthode de la sécante :
from scipy import optimize
# On définit la fonction à résoudre -> =0
fzero = lambda x:(Xi(x,0.9)/Psi(x,0.9)-f_a2)

sol = optimize.root_scalar(fzero, x0=1.4,x1=1.6, method='secant')
a2=sol.root
print('a2 : %.5f'%a2)
```

### 3) Déplacement radial de l'axe CD :

<div class="alert alert-info">
Le torseur des petits déplacements en un point $M$ d'un solide $S$ permet de représenter le champ des déplacements de $S$ pour un petit déplacement (rotation $\simeq 1$° et déplacement $\simeq 1$mm) de $S$ par rapport à un référentiel donné $\mathbb{R}_0$ :<br><br>
$$
\{\mathcal{D}_{S/\mathbb{R}_0}\}:
\left\lbrace
\begin{array}{c}
\vec{\Theta}_{S/\mathbb{R}_0}\\
\vec{U}(M)
\end{array}
\right\rbrace_{M}
$$
Avec :    
<ul>
    <li>$\vec{\Theta}_{S/\mathbb{R}_0}$ : Vecteur rotation de $S$ par rapport à $\mathbb{R}_0$.</li>
    <li>$\vec{U}(M)$ : Vecteur déplacement du point $M$ de $S$.</li>
</ul> 
Le torseur déplacement peut être défini de 2 manières :

Dans le cas de l'exercice on obtient les déplacements en $C$ et $D$ de $S_2$ par rapport à la position sans précontrainte et sans effort avec :

$$
\delta_{xi}=\delta_{ri}\frac{X_i}{\sqrt{X_i^2+Z_i^2}}
\quad
\delta_{yi}=0
\quad
\delta_{zi}=\delta_{ri}\frac{Z_i}{\sqrt{X_i^2+Z_i^2}}
\quad
i=\{1,2\}
$$

```python
dx1 = dr1*X1/np.sqrt(X1**2+Z1**2)
dy1 = 0
dz1 = dr1*Z1/np.sqrt(X1**2+Z1**2)

dx2 = dr2*X2/np.sqrt(X2**2+Z2**2)
dy2 = 0
dz2 = dr2*Z2/np.sqrt(X2**2+Z2**2)
print('En C:\n\tdx1=%+.6f mm\tdy1=%+.6f mm\tdz1=%+.6f mm\t'%(dx1,dy1,dz1))
print('En D:\n\tdx2=%+.6f mm\tdy2=%+.6f mm\tdz2=%+.6f mm\t'%(dx2,dy2,dz2))
```

Le torseur du champ des petits déplacements de $S_2$ par à $\mathbb{R}_0$ permet d'écrire :

$$
\{\mathcal{D}_{S_2/\mathbb{R}_0}\}:
\left\lbrace
\begin{array}{c}
\vec{\Theta}_{S_2/\mathbb{R}_0}\\
\vec{U}(D)
\end{array}
\right\rbrace_{D}
:
\left\lbrace
\begin{array}{c}
\vec{\Theta}_{S_2/\mathbb{R}_0}\\
\vec{U}(C)=\vec{U}(D)+\vec{CD}\wedge\vec{\Theta}_{S_2/\mathbb{R}_0}
\end{array}
\right\rbrace_{C}
$$

$$
\left\lbrace
\begin{array}{c}
\delta_{x1}\\
\delta_{y1}\\
\delta_{z1}
\end{array}
\right\rbrace
=
\left\lbrace
\begin{array}{c}
\delta_{x2}\\
\delta_{y2}\\
\delta_{z2}
\end{array}
\right\rbrace
+
\left\lbrace
\begin{array}{c}
0\\
-(L_1+L_2)\\
0
\end{array}
\right\rbrace
\wedge
\left\lbrace
\begin{array}{c}
\theta_x\\
\theta_y\\
\theta_z
\end{array}
\right\rbrace
$$

D'où l'expression des composantes du vecteur rotation :

$$
\theta_x = \frac{\delta_{z1}-\delta_{z2}}{L_1+L_2}\quad\theta_y=0\quad\theta_z = -\frac{\delta_{x1}-\delta_{x2}}{L_1+L_2}
$$

```python
thetax = (dz1-dz2)/(L1+L2)
thetaz = -(dx1-dx2)/(L1+L2)
print ('Thetax :%.5e rad\tThetaz :%.5e rad'%(thetax,thetaz))
```

Le vecteur rotation étant connu, on peut calculer le déplacement du point $O$ :

$$\vec{U}(O)=\vec{U}(D)+\vec{OD}\wedge\vec{\Theta}_{S_2/\mathbb{R}_0}$$

$$
\left\lbrace
\begin{array}{c}
\delta_{xO}\\
\delta_{yO}\\
\delta_{zO}
\end{array}
\right\rbrace
=
\left\lbrace
\begin{array}{c}
\delta_{x2}\\
\delta_{y2}\\
\delta_{z2}
\end{array}
\right\rbrace
+
\left\lbrace
\begin{array}{c}
0\\
-(d_1/2+L_2)\\
0
\end{array}
\right\rbrace
\wedge
\left\lbrace
\begin{array}{c}
\theta_x\\
\theta_y\\
\theta_z
\end{array}
\right\rbrace
=
\left\lbrace
\begin{array}{c}
\delta_{x2}-(d_1/2+L_2)\theta_z\\
\delta_{y2}\\
\delta_{z2}+(d_1/2+L_2)\theta_x
\end{array}
\right\rbrace
$$

$$\delta_{rO}=\sqrt{\delta_{xO}^2+\delta_{zO}^2}$$

```python
dxO = dx2 - (d1/2+L2)*thetaz
dzO = dx2 + (d1/2+L2)*thetax
drO = np.sqrt(dxO**2+dzO**2)
print('Dep. radial de O : %.5f mm'%drO)
```
